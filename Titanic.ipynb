# Use sns.pairplot(), sns.heatmap() for visualization

```python
sns.pairplot(
    df[['survived','age','fare','pclass','sibsp','parch']].dropna(),
    hue='survived',
    diag_kind='kde',
    corner=True
)
plt.savefig('pairplot', dpi=300, bbox_inches='tight')
plt.show()

plt.figure(figsize=(8,6))
sns.heatmap(df[['survived','pclass','age','sibsp','parch','fare']].corr(), 
            annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.savefig('Correlation Heatmap', dpi=300, bbox_inches='tight')
plt.show()
```

### Observations:
- Pairplot shows clear class separation in 'Fare' and 'Pclass' by survival.
- Heatmap: Fare is negatively correlated with Pclass; Family features (SibSp, Parch) show weak correlation.


# Identifing relationships and trends

```python
# Survival rate by Sex
sns.barplot(x='sex', y='survived', data=df, ci=None)
plt.title("Survival Rate by Sex")
plt.savefig('Survival Rate by Sex', dpi=300, bbox_inches='tight')
plt.show()

# Survival rate by Pclass
sns.barplot(x='pclass', y='survived', data=df, ci=None)
plt.title("Survival Rate by Pclass")
plt.savefig('Survival Rate by Pclass', dpi=300, bbox_inches='tight')
plt.show()

# Survival by Embarked
sns.barplot(x='embarked', y='survived', data=df, ci=None)
plt.title("Survival Rate by Embarked Port")
plt.savefig('Survival Rate by Embarked Port', dpi=300, bbox_inches='tight')
plt.show()
```

### Observations:
- Females survived at much higher rates than males.
- 1st class passengers had much higher survival than 3rd class.
- Passengers from port 'C' had higher survival than 'S'.


# Histograms, Boxplots, Scatterplots

```python
# Histogram: Age distribution
sns.histplot(df['age'].dropna(), bins=30, kde=True)
plt.title("Age Distribution")
plt.savefig('Age Distribution', dpi=300, bbox_inches='tight')
plt.show()

# Boxplot: Fare by Pclass
sns.boxplot(x='pclass', y='fare', data=df)
plt.title("Fare by Passenger Class")
plt.savefig('Fare by Passenger Class', dpi=300, bbox_inches='tight')
plt.show()

# Scatterplot: Age vs Fare colored by Survival
sns.scatterplot(x='age', y='fare', hue='survived', data=df)
plt.title("Age vs Fare by Survival")
plt.savefig('Age vs Fare by Survival', dpi=300, bbox_inches='tight')
plt.show()
```

### Observations:
- Age distribution is right-skewed with many young adults.
- Fare is much higher in 1st class with large variance (outliers).
- Survivors cluster at higher fares; some young children in 3rd class also survived.
